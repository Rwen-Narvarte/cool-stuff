```python
import numpy as np

board = np.empty((8,8))
board = board.astype(str)

#Officials
board[:] = "|__|"
board[:1] = ["(R1)","(H1)","(B1)","(KK)","(QQ)","(B2)","(H2)","(R2)"]

#Pawns dahil tamad magsulat
row1 = [ ]
for i in range(8):
	i = i + 1
	row1.append("(P" + str(i) + ")")
board[1:2] = row1

#Pawns dahil tamad magsulat
row6 = []
for i in range(8):
	i = i + 1
	row6.append("<P" + str(i) + ">")
board[6:7] = row6

#Officials ule
board[7:8] =  ["<R1>","<H1>","<B1>","<KK>","<QQ>","<B2>","<H2>","<R2>"]
print(board)



#Your move
counter = 1
game = True


def locator(mogus):
	if counter%2 == 1:
		x,y = np.where(board == "<" + str(mogus) + ">")
		return x[0],y[0]
	if counter%2 == 0:
		x,y = np.where(board == "(" + str(mogus) + ")")
		return x[0],y[0]
		
def translatetochess(x,y):
	cols = ["a","b","c","d","e","f","g","h"]
	rows = ["1","2","3","4","5","6","7","8"]
	return str(cols[int(x)] + rows[int(y)])

def translatetocode(xy):
	colcode = { "a" : 0, "b" : 1, "c":2,"d":3,"e":4,"f":5,"g":6,"h":7}
	rowcode = { "1":0,"2":1,"3":2,"4":3,"5":4,"6":5,"7":6,"8":7 }
	a,b = colcode[str(xy[0])] , rowcode[str(xy[1])]
	return a,b
	

#Pawn
def pawn(amogus):
	possiblemoves = []
	x,y = locator(amogus)
	possiblemoves.append(translatetochess(x,y+1))
	possiblemoves.append(translatetochess(x,y+2))
	

while game == True:
	move = input("kimi no move (use this format "P5e6"):")
	possiblemoves = []
	
	
	dict = { "P" : pawn }
	dict[move[:1]](move[:2])
	
	
	if move[2:] in possiblemoves:
		board[locator(move[:2])] = "|__|"
		if counter%2 ==1:
			board[translatetocode[move[2:]]] = str("<" + move[:2] + ">")
		else:
			board[translatetocode[move[2:]]] = str("(" + move[:2] + ")")
	
		
	counter = counter + 1
	print(board)

```
